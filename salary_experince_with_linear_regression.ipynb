```python
#step1:importing libraries
import pandas as pd
```

```python
#importing dataset
salary=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
#step3:define target (y) and features (x)
y=salary['Salary'] #independent
X=salary[['Experience Years']]  #dependent
```

```python
X.shape
```

```python
y.shape
```

```python
#step4: Train test Split
```

```python
from sklearn.model_selection import train_test_split
#Split arrays or matrices into random train and test subsets.
```

```python
#to divide data into 4 parts
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=2529)
X_train                  #random selection will be done upto 70 persent based on our requirement size
                             #default train-75% test-25%
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
#step5:select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#step6:train model(fit model)
model.fit(X_train,y_train)
```

```python
model.intercept_  #m
```

```python
model.coef_  #exp
```

```python
#step7:prediction
y_pred=model.predict(X_test) #bcz based on experience we are predicting, so X_test=experience ,so we are giving x_test
#predicting salary
```

```python
X_test
```

```python
#step8 Accuracy
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
